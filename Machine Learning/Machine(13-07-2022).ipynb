```python
df = pd.read_csv("Advertising.csv",index_col = [0])
df.head(2)
```

```python
plot_num = 1
plt.figure(figsize=(10,10))
for i in df.columns:
    if plot_num<5:
        plt.subplot(2,2,plot_num)
        st.probplot(df[i],dist = "norm",plot = pylab);
        plot_num+=1
```
